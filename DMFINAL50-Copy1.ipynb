```python
import pandas as pd
from itertools import combinations

# Load the filtered association rules from CSV
csv_filename = "filtered_association_rules.csv"
rules_df = pd.read_csv(csv_filename)

# Define minimum lift threshold
min_lift_threshold = 1.0  

# Step 1: Store all (From  To) mappings with Lift in a dictionary for faster lookup
two_itemset_lift = {(row["From"], row["To"]): row["Lift"] for _, row in rules_df.iterrows()}

# Step 2: Extract unique movies from association rules
movies = set(rules_df["From"]).union(set(rules_df["To"]))

# Step 3: Generate frequent 3-itemsets using only valid 2-itemsets
three_itemset_rules = []

for movie1, movie2, movie3 in combinations(movies, 3):
    # Check if the required 2-itemsets exist in our dictionary
    if (movie1, movie2) in two_itemset_lift and (movie2, movie3) in two_itemset_lift:
        lift1 = two_itemset_lift[(movie1, movie2)]
        lift2 = two_itemset_lift[(movie2, movie3)]

        # Compute average lift for the 3-itemset
        avg_lift = (lift1 + lift2) / 2

        # Store only if Lift > threshold
        if avg_lift > min_lift_threshold:
            three_itemset_rules.append({"Itemset": (movie1, movie2, movie3), "Lift": avg_lift})

# Convert to DataFrame
three_itemset_df = pd.DataFrame(three_itemset_rules)

# Sort by Lift value
three_itemset_df = three_itemset_df.sort_values(by="Lift", ascending=False)

# Save frequent 3-itemsets to CSV
three_itemset_filename = "frequent_3_itemsets.csv"
three_itemset_df.to_csv(three_itemset_filename, index=False)

# Print confirmation message
print(f"\nFrequent 3-itemsets (Lift > {min_lift_threshold}) saved to '{three_itemset_filename}'")

# Display top 10 frequent 3-itemsets
print("\nTop 10 Frequent 3-Itemsets:")
print(three_itemset_df.head(10))
```

```python
import pandas as pd
from itertools import permutations

# Load the frequent 3-itemsets from CSV
three_itemset_filename = "frequent_3_itemsets.csv"
three_itemset_df = pd.read_csv(three_itemset_filename)

# Define minimum confidence threshold
min_confidence_threshold = 0.5  

# Dictionary to store association rules
association_rules = []

# Generate association rules from 3-itemsets
for _, row in three_itemset_df.iterrows():
    itemset = eval(row["Itemset"]) if isinstance(row["Itemset"], str) else row["Itemset"]
    lift = row["Lift"]
    
    # Generate all possible rules from the 3-itemset
    for perm in permutations(itemset, 3):  
        A, B, C = perm  # Example: (A, B, C) means {A, B}  C
        
        # Approximate confidence using Lift
        confidence = lift / 2  
        
        # Store only strong rules
        if confidence > min_confidence_threshold:
            association_rules.append({
                "From": f"{{{A}, {B}}}",
                "To": C,
                "Lift": lift,
                "Confidence": round(confidence, 2)
            })

# Convert to DataFrame
rules_df = pd.DataFrame(association_rules)

# Sort by Confidence and Lift
rules_df = rules_df.sort_values(by=["Confidence", "Lift"], ascending=False)

# Save association rules to CSV
rules_filename = "generated_association_rules1.csv"
rules_df.to_csv(rules_filename, index=False)

# Print confirmation message
print(f"\nGenerated association rules saved to '{rules_filename}'")

# Display top 10 association rules
print("\nTop 10 Association Rules:")
print(rules_df.head(10))
```

```python
import pandas as pd

# Load generated association rules from CSV
rules_filename = "generated_association_rules1.csv"
rules_df = pd.read_csv(rules_filename)

def recommend_movies(watched_movies, top_n=5):
    """
    Recommend movies based on watched movies using association rules.
    """
    recommendations = {}

    # Iterate through rules and find matches with watched movies
    for _, row in rules_df.iterrows():
        antecedent = row["From"].strip("{}").split(", ")  # Extract {A, B}
        consequent = row["To"]
        confidence = row["Confidence"]
        lift = row["Lift"]

        # Check if the watched movies match the antecedent
        if set(antecedent).issubset(set(watched_movies)):  
            if consequent not in watched_movies:  # Recommend only unseen movies
                if consequent in recommendations:
                    recommendations[consequent].append((confidence, lift))
                else:
                    recommendations[consequent] = [(confidence, lift)]

    # Average confidence and lift for each recommended movie
    recommendations = {movie: (sum(c for c, _ in values) / len(values), sum(l for _, l in values) / len(values))
                       for movie, values in recommendations.items()}

    # Sort by Confidence & Lift
    sorted_recommendations = sorted(recommendations.items(), key=lambda x: (-x[1][0], -x[1][1]))

    # Get top N recommendations
    return [movie for movie, _ in sorted_recommendations[:top_n]]

# Example Usage
user_watched_movies = input("Enter movies you have watched (comma-separated): ").split(", ")
recommended_movies = recommend_movies(user_watched_movies)

# Display recommendations
if recommended_movies:
    print("\nRecommended movies for you:")
    for idx, movie in enumerate(recommended_movies, start=1):
        print(f"{idx}. {movie}")
else:
    print("No strong recommendations found.")
```

```python
import pandas as pd

# Load the association rules from CSV
csv_filename = "filtered_association_rules.csv"
rules_df = pd.read_csv(csv_filename)

# Function to recommend movies based on association rules
def recommend_movies(watched_movie, rules_df, top_n=5):
    """
    Given a watched movie, recommend top N associated movies.
    """
    # Find rules where the watched movie appears in the "From" column
    recommendations = rules_df[rules_df["From"] == watched_movie][["To", "Lift"]]
    
    if recommendations.empty:
        print(f"\nNo recommendations found for '{watched_movie}'")
        return []
    
    # Sort by Lift (strongest association first)
    recommendations = recommendations.sort_values(by="Lift", ascending=False)
    
    # Return top N recommended movies
    return recommendations["To"].head(top_n).tolist()

# Get user input
watched_movie = input("Enter a movie you watched: ")

# Get recommendations
recommended_movies = recommend_movies(watched_movie, rules_df)

# Print recommendations
if recommended_movies:
    print(f"\nBased on '{watched_movie}', we recommend these movies:")
    for idx, movie in enumerate(recommended_movies, start=1):
        print(f"{idx}. {movie}")
else:
    print("No recommendations available.")
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Load the movie ratings dataset (assumes columns: UserID, Movie, Rating)
ratings_df = pd.read_csv("movie_ratings.csv")  

# Create a user-movie matrix
user_movie_matrix = ratings_df.pivot_table(index="UserID", columns="Movie", values="Rating")

# Fill missing values with 0 (assumes unwatched movies have 0 rating)
user_movie_matrix = user_movie_matrix.fillna(0)

# Compute cosine similarity between users
user_similarity = cosine_similarity(user_movie_matrix)

# Convert to DataFrame
user_sim_df = pd.DataFrame(user_similarity, index=user_movie_matrix.index, columns=user_movie_matrix.index)

def recommend_movies(user_id, top_n=5):
    """
    Recommend movies for a given user based on collaborative filtering.
    """
    if user_id not in user_sim_df.index:
        print("User ID not found!")
        return []
    
    # Find the most similar users
    similar_users = user_sim_df[user_id].sort_values(ascending=False).index[1:]  # Exclude self
    
    # Get movies watched by the user
    watched_movies = user_movie_matrix.loc[user_id]
    watched_movies = watched_movies[watched_movies > 0].index  # Movies the user has rated
    
    recommendations = []
    
    # Go through similar users and recommend unseen movies
    for similar_user in similar_users:
        similar_user_movies = user_movie_matrix.loc[similar_user]
        
        # Recommend movies that the similar user has rated but target user hasn't seen
        for movie, rating in similar_user_movies.items():
            if movie not in watched_movies and rating > 3:  # Recommend highly rated unseen movies
                recommendations.append((movie, rating))

    # Sort recommendations by rating (highest first)
    recommendations = sorted(recommendations, key=lambda x: -x[1])[:top_n]

    return [movie for movie, _ in recommendations]

# Example usage: Get recommendations for a user
user_id = int(input("Enter User ID: "))  
recommended_movies = recommend_movies(user_id)

# Print recommendations
if recommended_movies:
    print(f"\nRecommended movies for User {user_id}:")
    for idx, movie in enumerate(recommended_movies, start=1):
        print(f"{idx}. {movie}")
else:
    print("No recommendations available.")
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Load movie ratings dataset
ratings_df = pd.read_csv("movie_ratings.csv")  # Ensure this file exists in the directory

# Pivot the data to create a user-movie matrix
user_movie_matrix = ratings_df.pivot_table(index="UserID", columns="Movie", values="Rating")

# Fill missing values with the mean rating of the user instead of 0 (better accuracy)
user_movie_matrix = user_movie_matrix.apply(lambda row: row.fillna(row.mean()), axis=1)

# Compute cosine similarity between users
user_similarity = cosine_similarity(user_movie_matrix.fillna(0))  # Handle NaN values properly

# Convert to DataFrame
user_sim_df = pd.DataFrame(user_similarity, index=user_movie_matrix.index, columns=user_movie_matrix.index)

def recommend_movies(user_id, top_n=5):
    """
    Recommend movies for a user based on similar users' ratings.
    """
    if user_id not in user_sim_df.index:
        print("User ID not found!")
        return []

    # Get similar users sorted by similarity score (excluding self)
    similar_users = user_sim_df[user_id].drop(user_id).sort_values(ascending=False).index

    # Get movies already watched by the user
    watched_movies = set(user_movie_matrix.loc[user_id].dropna().index)

    recommendations = {}

    # Aggregate ratings from similar users
    for similar_user in similar_users:
        similar_user_movies = user_movie_matrix.loc[similar_user].dropna()
        
        for movie, rating in similar_user_movies.items():
            if movie not in watched_movies and rating >= 4:  # Only recommend highly rated movies
                if movie in recommendations:
                    recommendations[movie].append(rating)
                else:
                    recommendations[movie] = [rating]

    # Average ratings for unseen movies
    recommendations = {movie: np.mean(ratings) for movie, ratings in recommendations.items()}

    # Get top N recommendations
    top_recommendations = sorted(recommendations.items(), key=lambda x: -x[1])[:top_n]

    return [movie for movie, _ in top_recommendations]

# Example Usage
try:
    user_id = int(input("Enter User ID: "))
    recommended_movies = recommend_movies(user_id)

    if recommended_movies:
        print(f"\nRecommended movies for User {user_id}:")
        for idx, movie in enumerate(recommended_movies, start=1):
            print(f"{idx}. {movie}")
    else:
        print("No recommendations available.")
except ValueError:
    print("Invalid User ID. Please enter a valid integer.")
```
